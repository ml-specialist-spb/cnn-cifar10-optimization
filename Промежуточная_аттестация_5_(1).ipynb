```python
# Модуль 5: Разработка и оптимизация CNN на CIFAR-10 — Colab-ready
# Этапы: 1) базовая модель;
# 2) +BatchNorm;
# 3) +Dropout;
# 4) ядро 5x5
# В конце: сводная таблица, графики, и доп. блок (*) — Transfer Learning (MobileNetV2)

import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
from tensorflow.keras import layers, models

# Воспроизводимость
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

print("Доступные GPU:", tf.config.list_physical_devices('GPU'))

# 1) Загрузка CIFAR-10
from tensorflow.keras.datasets import cifar10
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
num_classes = 10
class_names = ['airplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck']

# One-hot для меток
y_train_cat = tf.keras.utils.to_categorical(y_train, num_classes)
y_test_cat  = tf.keras.utils.to_categorical(y_test, num_classes)

# Общие гиперпараметры
EPOCHS = 30         # можно увеличить/уменьшить, EarlyStopping остановит раньше
BATCH_SIZE = 128
VAL_SPLIT = 0.1

early_stop = tf.keras.callbacks.EarlyStopping(
    monitor='val_accuracy', patience=5, restore_best_weights=True
)
reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss', factor=0.5, patience=2, min_lr=1e-5, verbose=1
)

# Вспомогательный конструктор сверточного блока
def conv_block(filters, kernel_size=(3,3), use_bn=False):
    # Возвращает список слоев: Conv2D -> (BN -> ReLU) или Conv2D(activation='relu')
    blk = []
    if use_bn:
        blk.append(layers.Conv2D(filters, kernel_size, padding='same', use_bias=False))
        blk.append(layers.BatchNormalization())
        blk.append(layers.ReLU())
    else:
        blk.append(layers.Conv2D(filters, kernel_size, padding='same', activation='relu'))
    return blk

# Построение модели по параметрам
def build_cnn(input_shape=(32,32,3), kernel_size=(3,3), use_bn=False, use_dropout=False, dropout_rate=0.5):
    inputs = layers.Input(shape=input_shape)
    x = layers.Rescaling(1./255)(inputs)

    # Блок 1
    for lyr in conv_block(32, kernel_size, use_bn): x = lyr(x)
    for lyr in conv_block(32, kernel_size, use_bn): x = lyr(x)
    x = layers.MaxPooling2D()(x)
    x = layers.Dropout(0.25)(x)  # легкая регуляризация фич-карт

    # Блок 2
    for lyr in conv_block(64, kernel_size, use_bn): x = lyr(x)
    for lyr in conv_block(64, kernel_size, use_bn): x = lyr(x)
    x = layers.MaxPooling2D()(x)
    x = layers.Dropout(0.25)(x)

    # Блок 3
    for lyr in conv_block(128, kernel_size, use_bn): x = lyr(x)
    for lyr in conv_block(128, kernel_size, use_bn): x = lyr(x)
    x = layers.MaxPooling2D()(x)
    x = layers.Dropout(0.25)(x)

    x = layers.Flatten()(x)
    x = layers.Dense(256, activation='relu')(x)
    if use_dropout:
        x = layers.Dropout(dropout_rate)(x)  # Dropout после Dense по условию этапа 3

    outputs = layers.Dense(num_classes, activation='softmax')(x)
    model = models.Model(inputs, outputs)
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Обучение и оценка
def train_and_eval(model, name):
    print(f"\n=== Обучение: {name} ===")
    hist = model.fit(
        X_train, y_train_cat,
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        validation_split=VAL_SPLIT,
        callbacks=[early_stop, reduce_lr],
        verbose=1
    )
    test_loss, test_acc = model.evaluate(X_test, y_test_cat, verbose=0)
    print(f"{name} — Точность на тесте: {test_acc:.4f}")
    return hist, test_acc

# Графики обучения (accuracy и loss)
def plot_history(hist, title):
    h = hist.history
    fig, ax = plt.subplots(1, 2, figsize=(11,4))
    ax[0].plot(h['accuracy'], label='train')
    ax[0].plot(h['val_accuracy'], label='val')
    ax[0].set_title(f'{title} — Accuracy'); ax[0].set_xlabel('Epoch'); ax[0].grid(True); ax[0].legend()

    ax[1].plot(h['loss'], label='train')
    ax[1].plot(h['val_loss'], label='val')
    ax[1].set_title(f'{title} — Loss'); ax[1].set_xlabel('Epoch'); ax[1].grid(True); ax[1].legend()

    plt.tight_layout()
    plt.show()

# Этап 1: базовая модель (ядра 3x3, без BN и Dropout после Dense)
model_1 = build_cnn(kernel_size=(3,3), use_bn=False, use_dropout=False)
hist_1, acc_1 = train_and_eval(model_1, "Этап 1 — Базовая 3x3")
plot_history(hist_1, "Этап 1 — Базовая 3x3")

# Этап 2: добавляем Batch Normalization (после каждого Conv)
model_2 = build_cnn(kernel_size=(3,3), use_bn=True, use_dropout=False)
hist_2, acc_2 = train_and_eval(model_2, "Этап 2 — +BatchNorm 3x3")
plot_history(hist_2, "Этап 2 — +BatchNorm 3x3")

# Этап 3: добавляем Dropout после Dense
model_3 = build_cnn(kernel_size=(3,3), use_bn=True, use_dropout=True, dropout_rate=0.5)
hist_3, acc_3 = train_and_eval(model_3, "Этап 3 — +BatchNorm +Dropout 3x3")
plot_history(hist_3, "Этап 3 — +BatchNorm +Dropout 3x3")

# Этап 4: меняем размер ядер на 5x5 (оставляем BN и Dropout)
model_4 = build_cnn(kernel_size=(5,5), use_bn=True, use_dropout=True, dropout_rate=0.5)
hist_4, acc_4 = train_and_eval(model_4, "Этап 4 — +BatchNorm +Dropout 5x5")
plot_history(hist_4, "Этап 4 — +BatchNorm +Dropout 5x5")

# Сводная таблица
results = pd.DataFrame([
    {"Этап": 1, "Архитектура": "Базовая 3x3", "Accuracy (test)": acc_1},
    {"Этап": 2, "Архитектура": "+BatchNorm 3x3", "Accuracy (test)": acc_2},
    {"Этап": 3, "Архитектура": "+BatchNorm +Dropout 3x3", "Accuracy (test)": acc_3},
    {"Этап": 4, "Архитектура": "+BatchNorm +Dropout 5x5", "Accuracy (test)": acc_4},
]).sort_values("Этап")
print("\nСводная таблица результатов:")
print(results)

# Пример предсказания и визуализации
def show_sample_prediction(model, idx=0):
    img = X_test[idx]
    probs = model.predict(img[None, ...], verbose=0)[0]
    pred = int(np.argmax(probs))
    plt.imshow(img)
    plt.title(f"Pred: {class_names[pred]}  p={probs[pred]:.2f}")
    plt.axis('off'); plt.show()

print("\nПример предсказания (Этап 3 — как обычно сильный):")
show_sample_prediction(model_3, idx=0)

# Краткие выводы (автоматический черновик)
best_row = results.iloc[results['Accuracy (test)'].idxmax()]
print("\nВыводы:")
print(f"- Требование 60% выполнено? {'Да' if results['Accuracy (test)'].max() >= 0.60 else 'Нет'}")
print("- Этап 1 — базовая модель задает базовую точку;")
print("- Этап 2 — BatchNorm обычно ускоряет сходимость и стабилизирует обучение;")
print("- Этап 3 — Dropout после Dense помогает бороться с переобучением;")
print("- Этап 4 — смена ядра на 5x5 может улучшить/ухудшить в зависимости от данных и регуляризации;")
print(f"- Лучшая модель: Этап {int(best_row['Этап'])} ({best_row['Архитектура']}) с точностью {best_row['Accuracy (test)']:.4f}")
```

```python
# Доп. задание: Transfer Learning (MobileNetV2) — standalone Colab-ready код
# Загружает CIFAR-10, строит модель на базе MobileNetV2 (замороженная база),
# обучает, оценивает и строит графики обучения.

import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
import matplotlib.pyplot as plt

# Воспроизводимость
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

print("Доступные GPU:", tf.config.list_physical_devices('GPU'))

# Загрузка CIFAR-10
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
num_classes = 10
y_train_cat = to_categorical(y_train, num_classes)
y_test_cat  = to_categorical(y_test, num_classes)

# Гиперпараметры
EPOCHS = 8
BATCH_SIZE = 128
VAL_SPLIT = 0.1

early_stop = tf.keras.callbacks.EarlyStopping(
    monitor='val_accuracy', patience=5, restore_best_weights=True
)
reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss', factor=0.5, patience=2, min_lr=1e-5, verbose=1
)

# Вспомогательная функция для графиков
def plot_history(hist, title=""):
    h = hist.history
    fig, ax = plt.subplots(1, 2, figsize=(12,4))
    ax[0].plot(h.get('accuracy', []), label='train')
    ax[0].plot(h.get('val_accuracy', []), label='val')
    ax[0].set_title(f'{title} — Accuracy'); ax[0].set_xlabel('Epoch'); ax[0].grid(True); ax[0].legend()

    ax[1].plot(h.get('loss', []), label='train')
    ax[1].plot(h.get('val_loss', []), label='val')
    ax[1].set_title(f'{title} — Loss'); ax[1].set_xlabel('Epoch'); ax[1].grid(True); ax[1].legend()

    plt.tight_layout()
    plt.show()

# Построение модели трансферного обучения
def build_tl_model(input_shape=(32,32,3), tl_input_size=128, train_base=False):
    """
    input_shape: форма исходных изображений (CIFAR-10 -> 32x32x3)
    tl_input_size: размер для MobileNetV2 (например, 128 -> 128x128)
    train_base: если True - размораживаем базовую модель для дообучения
    """
    inputs = layers.Input(shape=input_shape)
    # Меняем размер до того, что ожидает предобученная модель
    x = layers.Resizing(tl_input_size, tl_input_size)(inputs)
    # Препроцессинг для MobileNetV2
    x = layers.Lambda(preprocess_input)(x)
    # Подключаем базовую модель, передаём tensor x как input_tensor
    base = MobileNetV2(include_top=False, weights='imagenet', input_tensor=x)
    base.trainable = train_base

    x = layers.GlobalAveragePooling2D()(base.output)
    x = layers.Dense(256, activation='relu')(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)

    model = models.Model(inputs, outputs)
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Строим модель (замороженная база)
tl_model = build_tl_model(input_shape=(32,32,3), tl_input_size=128, train_base=False)
tl_model.summary()

# Обучение
tl_hist = tl_model.fit(
    X_train, y_train_cat,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    validation_split=VAL_SPLIT,
    callbacks=[early_stop, reduce_lr],
    verbose=1
)

# Оценка
tl_loss, tl_acc = tl_model.evaluate(X_test, y_test_cat, verbose=0)
print(f"Transfer Learning (MobileNetV2, frozen base) — test accuracy: {tl_acc:.4f}, loss: {tl_loss:.4f}")

# Графики обучения
plot_history(tl_hist, "Transfer Learning — MobileNetV2")

# Пример предсказания
import numpy as np
import matplotlib.pyplot as plt
class_names = ['airplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck']

def show_sample_prediction(model, X, idx=0):
    img = X[idx]
    probs = model.predict(img[None, ...], verbose=0)[0]
    pred = int(np.argmax(probs))
    plt.imshow(img.astype('uint8'))
    plt.title(f"Pred: {class_names[pred]}  p={probs[pred]:.2f}")
    plt.axis('off'); plt.show()

show_sample_prediction(tl_model, X_test, idx=0)
```
